**Calculatiing Average Profit Margins**

To determine the profit margin per sale accurately, wee need to

1. calculate the ***profit*** for each sale by subtracting the ***state _bottle_cost*** from the ***state_bottle_price***.  
2. determine the profit margin divide the ***profit*** by the ***state_bottle_price***. Then multiply the result by ***100*** to get a percentage value.
3. Finally, determine the ***profit margin*** by utilizing the ***mean() function***.


```python
# Calculating the profits
cleaned_iowa_liquor_sales['profit'] = cleaned_iowa_liquor_sales['state_bottle_retail'] - cleaned_iowa_liquor_sales['state_bottle_cost']

# Calculating the profit margin
cleaned_iowa_liquor_sales['profit_margin'] = (cleaned_iowa_liquor_sales['profit'] / cleaned_iowa_liquor_sales['state_bottle_retail']) * 100


# Calculating Average Profit Margin using mean()
cleaned_iowa_liquor_sales['profit_margin'].mean()
```

**Top Selling Vodkas, Whiskies, and Flavored liquours**

```python
# Filter for Vodka, Whiskey, and flavored liquors
vodka_df = cleaned_iowa_liquor_sales[cleaned_iowa_liquor_sales['category_name'].str.contains('Vodka', case=False)]
whiskey_df = cleaned_iowa_liquor_sales[cleaned_iowa_liquor_sales['category_name'].str.contains('Whiskey', case=False)]
flavored_liquor_df = cleaned_iowa_liquor_sales[cleaned_iowa_liquor_sales['category_name'].str.contains('Liqueur|Cordial', case=False)]

# Group by year and item description, then sum the sales volume
vodka_top_by_year = vodka_df.groupby(['year', 'item_description'])['sale_dollars'].sum().groupby(level=0).nlargest(3)
whiskey_top_by_year = whiskey_df.groupby(['year', 'item_description'])['sale_dollars'].sum().groupby(level=0).nlargest(3)
flavored_liquor_top_by_year = flavored_liquor_df.groupby(['year', 'item_description'])['sale_dollars'].sum().groupby(level=0).nlargest(3)

# Convert the results to DataFrames for better formatting
vodka_top_by_year_df = vodka_top_by_year.reset_index(level=0, drop=True).reset_index()
whiskey_top_by_year_df = whiskey_top_by_year.reset_index(level=0, drop=True).reset_index()
flavored_liquor_top_by_year_df = flavored_liquor_top_by_year.reset_index(level=0, drop=True).reset_index()

# Set column names for clarity
vodka_top_by_year_df.columns = ['Year', 'Item Description', 'Sales in Dollars']
whiskey_top_by_year_df.columns = ['Year', 'Item Description', 'Sales in Dollars']
flavored_liquor_top_by_year_df.columns = ['Year', 'Item Description', 'Sales in Dollars']

# Function to print formatted tables with separation by year
def print_table(df, title):
    print(f"\n{title}")
    print()
    print(f"{'Year':<6} | {'Item Description':<40} | {'Sales in Dollars':>15}")
    print("-" * 70)
    current_year = None

    # Iterate through DataFrame and print each row
    for _, row in df.iterrows():
        if row['Year'] != current_year:
            # Add a line break between different years
            print("\n" + "-" * 70)
            current_year = row['Year']
        print(f"{row['Year']:<6} | {row['Item Description']:<40} | {row['Sales in Dollars']:>15.2f}")

# Print results for Vodka
print_table(vodka_top_by_year_df, "Top 3 Vodkas by Year")

# Print results for Whiskey
print_table(whiskey_top_by_year_df, "Top 3 Whiskies by Year")

# Print results for Flavored Liquor
print_table(flavored_liquor_top_by_year_df, "Top 3 Flavored Liquors by Year")


```

```python
# Calculate Q1, Q3, and IQR for each numerical variable
def find_outliers(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  return outliers

# Find outliers for all numerical variables
outlier_counts = {}
for col in df.select_dtypes(include=['number']).columns:
  outliers = find_outliers(df, col)
  outlier_counts[col] = len(outliers)

# Sort the outlier counts in descending order
sorted_outliers = dict(sorted(outlier_counts.items(), key=lambda item: item[1], reverse=True))

# Get the top two variables with the highest number of outliers
top_two_variables = list(sorted_outliers.keys())[:2]

print("Top two variables with the highest number of outliers:", top_two_variables)

# Create box plots to verify the outliers
plt.figure(figsize=(12, 6))
for i, col in enumerate(top_two_variables, 1):
  plt.subplot(1, 2, i)
  plt.boxplot(df[col])
  plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

To build a linear regression model, it is crucial to first identify the target variable (the outcome you want to predict) and the features (the factors used to make the prediction). The target variable, denoted as \(y\), represents the dependent variable. In the case of the `cleaned_iowa_liquor_sales` dataset, possible target variables include **'Bottles Sold'**, **'Sale (Dollars)'**, or **'Volume Sold (Liters)'**, as these reflect key outcomes that could be influenced by other factors in the dataset.

Once the target variable is established, you need to determine the independent variables, or features, that will be used to make predictions. Features (\(X\)) are the factors believed to have an influence on the target variable. Potential numerical features in the dataset include **'Pack'**, **'Bottle Volume (ml)'**, **'State Bottle Cost'**, **'State Bottle Retail'**, and even **'Year'**, depending on how it's treated. Additionally, there are categorical features such as **'Zip Code'**, **'County'**, **'Category Name'**, **'Vendor Name'**, and **'Item Description'**. However, these categorical features cannot be directly used in a linear regression model and need to be converted into numerical form through a process called **dummy coding**.

Dummy coding, or **one-hot encoding**, creates binary (0 or 1) columns for each unique category within a categorical variable. For example, the **'County'** feature would be transformed into several columns like **'County_Polk'**, **'County_Linn'**, and **'County_Scott'**. If a record belongs to Polk County, the **'County_Polk'** column will be assigned a value of 1, while the other county columns will have values of 0. This process allows the model to use categorical data effectively.

After applying dummy coding to categorical variables, the next step is selecting the most relevant features to use in the model. This involves choosing the variables that have the strongest relationship with the target variable to improve the accuracy of predictions.

**Elbow Method:**

The goal here is to plot the sum of squared distances (inertia) for different values of K (number of clusters) and find the point where the decrease in inertia slows down significantly (the “elbow”). This point suggests the optimal number of clusters.

The Elbow method is simpler and more intuitive.

```python

# Select only numerical columns
numerical_cols = df_filtered.select_dtypes(include=['float64', 'int64']).columns
X_numerical = df_filtered[numerical_cols]

# Elbow Method to find optimal number of clusters
inertia = []
K_range = range(1, 11)  # Testing K values from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_numerical)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, 'bo-', markersize=8)
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia (Sum of squared distances)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

The “elbow” appears around **K=3**. This is the point where the rate of decrease in the inertia (sum of squared distances) slows down significantly. After **K=3**, the reduction in inertia becomes more gradual, indicating diminishing returns for adding more clusters.

Therefore, based on the elbow method, **3 clusters** seems to be the optimal choice.

```python
from sklearn.metrics import silhouette_score
```

```python
# Silhouette Analysis to find optimal number of clusters
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_numerical)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_numerical, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the Silhouette Analysis graph
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, 'bo-', markersize=8)
plt.xlabel('Number of clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()
```

The **Silhouette Analysis** plot shows that the highest silhouette score occurs at K=5, which suggests that 5 clusters might be the optimal number for this data indicating the clusters are well separated and compact at this number.

And the **Elbow Method** suggested K=3 as a reasonable choice, but there was still some ambiguity as inertia continued to decrease.

The choice for **optimal number** for the clustering model is ***K=5*** for the clustering model since it provides better-defined clusters according to the silhouette score.


```python
# Create K-means Clustering Model with K=5

from sklearn.cluster import KMeans

# Create K-means model with 5 clusters and a fixed random seed for reproducibility
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model to the numerical dataset
kmeans.fit(X_numerical)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Append the cluster labels to the original dataset
df_filtered['Clusters'] = cluster_labels

# Show the first few rows to confirm the cluster labels have been added
df_filtered.head()
```
